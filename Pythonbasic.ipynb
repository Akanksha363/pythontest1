6. Write a function to demonstrate the behavior of mutable and immutable arguments.

```python
def modify_arguments(number, my_list):
    number += 1  # Modifying the immutable argument (number)
    my_list.append(5)  # Modifying the mutable argument (my_list)

# Example usage
num =15
my_list = [1, 2, 3,4]

modify_arguments(num, my_list)

print(num)       # Output: 10 (original value remains unchanged)
print(my_list)   # Output: [1, 2, 3, 4,5] (list is modified)
```

9.Write a Python program to convert User input from string to integer, float,
and boolean types.

```python
def convert_input():
  """Converts user input to integer, float, and boolean types."""

  user_input = input("Enter a value: ")

  try:
    # convert to integer
    integer_value = int(user_input)
    print("Integer:", integer_value)
  except ValueError:
    print("Could not convert to integer.")

  try:
    # convert to float
    float_value = float(user_input)
    print("Float:", float_value)
  except ValueError:
    print("Could not convert to float.")

  # convert to boolean
  if user_input.lower() in ["true", "1"]:
    boolean_value = True
  elif user_input.lower() in ["false", "0"]:
    boolean_value = False
  else:
    boolean_value = None  # Or handle the case as needed

  if boolean_value is not None:
    print("Boolean:", boolean_value)
  else:
    print("Could not convert to boolean.")

# Example usage
convert_input()
```

10. Write code to demonstrate typecasting with list elements.

```python
# Demonstrating Type Casting with List Elements

mixed_list = ["1", "2", "3", "4.5", "5.6"]

print("Original List:")
print(mixed_list)


int_list = [int(float(item)) for item in mixed_list]
# Convert each element to a float
float_list = [float(item) for item in mixed_list]

# Convert each element to a string
str_list = [str(item) for item in mixed_list]

# Display the results
print("\nAfter Type Casting:")

print("List with elements cast to integers:")
print(int_list)

print("\nList with elements cast to floats:")
print(float_list)

print("\nList with elements cast to strings:")
print(str_list)
```

11. Write a program that checks if a number is positive, negative, or Zero.

```python
def check_number(number):
    """Checks if a number is positive, negative, or zero."""

    if number > 0:
        print(f"{number} is positive.")
    elif number < 0:
        print(f"{number} is negative.")
    else:
        print(f"{number} is zero.")

# Get user input
number = float(input("Enter a number: "))

# Call the function to check the number
check_number(number)
```

14. Write a program to Reverse a string using a While Loop.

```python
# Program to reverse a string using a while loop


original_string = input("Enter a string to reverse: ")


reversed_string = ""
index = len(original_string) - 1


while index >= 0:
    reversed_string += original_string[index]

    index -= 1

# Output the reversed string
print("Reversed String:", reversed_string)
```

15. Write a Python program to Calculate the factorial of a number provided by the user using a while Loop.

```python
num = int(input("Enter a number: "))
factorial = 1
i = num

if num < 0:
    print("Factorial is not defined for negative numbers.")
elif num == 0:
    print("The factorial of 0 is 1.")
else:
    while i > 0:
        factorial *= i
        i -= 1
    print(f"The factorial of {num} is {factorial}.")
```
